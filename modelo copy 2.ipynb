## Modelo de Previsão de Sucesso de Startups
---
**Desafio**: Imagine que você foi contratado por uma das maiores aceleradoras do mundo. Sua missão? Desenvolver um modelo preditivo capaz de identificar quais startups apresentam maior probabilidade de se tornarem casos de sucesso no mercado.

**Solução Esperada**: Modelo de machine learning

**Para quem a Solução se Destina**: Investidores e aceleradores na tomada de decisões mais estratégicas

**Dados disponibilizados**: 
- train.csv → conjunto de treinamento com startups + variável de destino (rótulos)
- test.csv → conjunto de teste (sem a coluna de destino)
- sample_submission.csv → formato de submissão esperado

**Metodologia**:

Para desenvolver a solução, a construção de um modelo de previsão de sucesso de startups, é necessário seguir uma sequência de etapas que organize o processo de desenvolvimento. O objetivo é mapear o problema de forma estruturada e compreender qual abordagem de fato poderá solucioná-lo.

Será utilizada a metodologia **CRISP-DM**, adaptada ao desafio. Diferente de um contexto empresarial tradicional, o foco aqui está no entendimento do **problema do desafio**. As etapas são:

1. **Entendimento do contexto** – definição e análise do desafio.  
2. **Exploração dos dados** – levantamento inicial de padrões e inconsistências.  
3. **Preparação dos dados** – limpeza, transformação e formatação adequada.  
4. **Modelagem** – aplicação de algoritmos para previsão de sucesso.  
5. **Avaliação** – análise de métricas e resultados obtidos.  

Ao final, em vez de uma implantação no ambiente de negócio, será realizada a **submissão da solução ao desafio**.


### 1. Entendimento do contexto 

Nesta atividade, cada participante terá acesso ao mesmo conjunto de dados estruturados, representando um problema de previsão real.
O desafio é desenvolver modelos capazes de gerar previsões precisas, que serão avaliadas de acordo com uma métrica específica definida pelos instrutores .

Avaliação:
A avaliação será realizada automaticamente pela plataforma Kaggle, com base nos padrões de submissão definidos no concurso.

Atenção: 
é essencial que todas as submissões sigam rigorosamente o formato exigido, pois qualquer desvio poderá resultar em erros ou desclassificação .

Em situações específicas (como inconsistências, problemas técnicos ou suspeitas de irregularidades), a avaliação pode ser revisada manualmente pelos instrutores.

4) Criação de pipelines para tratamento dos dados 
- Preenchimento de nan e criação de colunas binárias que indica se tinha ou n algum dado

```python
# ================================
# Criação dos pipelines de pré-processamento
# ================================

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Seleção das colunas age_*
age_cols = [
    "age_first_funding_year", 
    "age_last_funding_year",
    "age_first_milestone_year", 
    "age_last_milestone_year"
]

# Pipeline para imputar + criar indicador de faltantes
age_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median", add_indicator=True))
])

# Aplicação apenas nessas colunas para ver resultado
X_age = age_pipeline.fit_transform(X[age_cols])

# Recuperar nomes das colunas (originais + indicadores)
feat_names_age = age_pipeline.named_steps["imputer"].get_feature_names_out(age_cols)

import pandas as pd
X_age_df = pd.DataFrame(X_age, columns=feat_names_age, index=X.index)

display(X_age_df.head(10))

# ================================
# Preparação das outras colunas numéricas
# ================================

# Colunas numéricas (sem incluir funding_total_usd, que é tratado separadamente)
other_num_cols = [
    "relationships", "funding_rounds", 
    "milestones", "avg_participants"
]

other_num_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

# Teste em apenas algumas colunas
X_other_num = other_num_pipeline.fit_transform(X[other_num_cols])

import pandas as pd
X_other_num_df = pd.DataFrame(X_other_num, columns=other_num_cols, index=X.index)

display(X_other_num_df.head(10))

# ================================
# Preparação da coluna funding_total_usd
# ================================

# 1) Transformador simples para clip por quantis
class QuantileClipper(BaseEstimator, TransformerMixin):
    def __init__(self, lower=0.01, upper=0.99):
        self.lower = lower
        self.upper = upper
        self.low_ = None
        self.high_ = None

    def fit(self, X, y=None):
        X_ = np.asarray(X, dtype=float).reshape(-1, 1)
        self.low_  = np.nanpercentile(X_, self.lower * 100)
        self.high_ = np.nanpercentile(X_, self.upper * 100)
        return self

    def transform(self, X):
        X_ = np.asarray(X, dtype=float).reshape(-1, 1)
        return np.clip(X_, self.low_, self.high_)
    
    def get_feature_names_out(self, input_features=None):
        if input_features is None:
            # fallback genérico do sklearn
            return np.array(["x0"])
        return np.asarray(input_features)

# 2) Pipeline da coluna funding_total_usd
funding_col = ["funding_total_usd"]

funding_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("clipper", QuantileClipper(lower=0.01, upper=0.99)),  # winsoriza
    ("scaler", RobustScaler())                              # escala robusta
])

# 3) Aplicar e inspecionar
X_funding = funding_pipeline.fit_transform(X[funding_col])

X_funding_df = pd.DataFrame(
    X_funding, columns=[f"{funding_col[0]}_robust"], index=X.index
)

display(X[funding_col].describe())        # antes
display(X_funding_df.describe())          # depois
X_funding_df.head(10)

# ================================
# Montagem do ColumnTransformer final
# ================================

# 1) Binárias: imputar 0 (só para garantir ausência vira 0)
binary_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value=0))
])

# 2) Categóricas: imputar "MISSING" + OneHot
cat_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),  # "MISSING" tb funciona, mas most_frequent costuma ir bem
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

# 3) Montar o ColumnTransformer com TODOS os blocos
transformers = []

if age_cols:
    transformers.append(("age", age_pipeline, age_cols))

if other_num_cols:
    transformers.append(("num", other_num_pipeline, other_num_cols))

if funding_col:
    transformers.append(("funding", funding_pipeline, funding_col))

if binary_cols:
    transformers.append(("bin", binary_pipeline, binary_cols))

if cat_cols:
    transformers.append(("cat", cat_pipeline, cat_cols))

preprocess = ColumnTransformer(
    transformers=transformers,
    remainder="drop"  # mantém só o que foi mapeado nos transformers
)

X_pp = preprocess.fit_transform(X)  

print("Shape antes:", X.shape)
print("Shape depois:", X_pp.shape)

preprocess.fit(X)

feat_names = preprocess.get_feature_names_out()
len(feat_names), feat_names[:20]
```

### 4) Modelagem 

```python
# Imports
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    HistGradientBoostingClassifier,
    VotingClassifier
)
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.pipeline import Pipeline
import pandas as pd
import numpy as np

# 1) Separação X e y em cada base
y_train = df_train["labels"].copy()
X_train = df_train.drop(columns=["labels", "id"], errors="ignore")

X_test = df_test.drop(columns=["labels", "id"], errors="ignore")
X_test = X_test.reindex(columns=X_train.columns, fill_value=np.nan)

# 2) Ensemble (soft voting) com três estimadores
rf = RandomForestClassifier(
    n_estimators=800, max_depth=None,
    n_jobs=-1, random_state=42
)
et = ExtraTreesClassifier(
    n_estimators=800, max_depth=None,
    n_jobs=-1, random_state=42
)
hgb = HistGradientBoostingClassifier(
    random_state=42
)

voter = VotingClassifier(
    estimators=[("rf", rf), ("et", et), ("hgb", hgb)],
    voting="soft"
)

pipe_ens = Pipeline(steps=[
    ("preprocess", preprocess),
    ("clf", voter)
])

# 3) Espaço de busca (enxuto e efetivo)
param_grid_ens = {
    # pesos do ensemble
    "clf__weights": [(1,1,1), (2,1,2), (1,2,2), (2,2,1), (1,1,2)],
    # ajustes leves em cada base learner
    "clf__rf__min_samples_leaf": [1, 2, 4],
    "clf__rf__max_features": ["sqrt", "log2", None],
    "clf__et__min_samples_leaf": [1, 2, 4],
    "clf__et__max_features": ["sqrt", "log2", None],
    "clf__hgb__learning_rate": [0.03, 0.06, 0.1],
    "clf__hgb__max_leaf_nodes": [31, 63, 127],
    "clf__hgb__l2_regularization": [0.0, 1e-3, 1e-2],
    "clf__hgb__early_stopping": [True],
    "clf__hgb__validation_fraction": [0.1],
    "clf__hgb__n_iter_no_change": [20],
}

# 4) Validação e métricas
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {
    "accuracy": "accuracy",
    "recall_macro": "recall_macro",
    "f1_macro": "f1_macro",
}

grid_ens = GridSearchCV(
    estimator=pipe_ens,
    param_grid=param_grid_ens,
    scoring=scoring,
    refit="accuracy",   # foco em acurácia
    cv=cv,
    n_jobs=-1,
    verbose=1
)

# 5) Treino
grid_ens.fit(X_train, y_train)

print("[ENS] best params (by accuracy):", grid_ens.best_params_)
best_idx = grid_ens.best_index_
print("[ENS] CV accuracy:", grid_ens.cv_results_["mean_test_accuracy"][best_idx])
print("[ENS] CV recall_macro:", grid_ens.cv_results_["mean_test_recall_macro"][best_idx])
print("[ENS] CV f1_macro:", grid_ens.cv_results_["mean_test_f1_macro"][best_idx])

# 6) Previsão e submissão
y_pred_ens = grid_ens.predict(X_test)
pd.DataFrame({"id": df_test["id"], "labels": y_pred_ens}).to_csv("submissao_ensemble.csv", index=False)
```
