```python
# Organizando as colunas por tipo de dado

num_cols = [  
    "age_first_funding_year", "age_last_funding_year",
    "age_first_milestone_year", "age_last_milestone_year",
    "relationships", "funding_rounds",
    "funding_total_usd", "milestones",
    "avg_participants"
]
binary_cols = [  
    "is_CA","is_NY","is_MA","is_TX","is_otherstate",
    "is_software","is_web","is_mobile","is_enterprise","is_advertising",
    "is_gamesvideo","is_ecommerce","is_biotech","is_consulting","is_othercategory",
    "has_VC","has_angel","has_roundA","has_roundB","has_roundC","has_roundD"
]
cat_cols = ["category_code"]  
id_cols = ["id"]  
target_col = "labels"

# Separar alvo e features
y = df_train[target_col].copy()
X = df_train.drop(columns=[target_col] + [c for c in id_cols if c in df_train.columns])  

# Manter nas listas apenas colunas que realmente existem em X
def keep_existing(cols, X):
    return [c for c in cols if c in X.columns]

num_cols   = keep_existing(num_cols, X)  
binary_cols= keep_existing(binary_cols, X)  
cat_cols   = keep_existing(cat_cols, X)  

# Checagens 
print("X shape:", X.shape, "| y shape:", y.shape) 
print(f"Numéricas: {len(num_cols)}  | Binárias: {len(binary_cols)}  | Categóricas: {len(cat_cols)}") 
print("Primeiras colunas de X:", list(X.columns)[:8]) 
display(X.head(3))
y.value_counts(normalize=True).mul(100).round(2)
```

### 3. Criação de pipelines para tratamento dos dados 

```python
# ================================
# PRÉ-PROCESSAMENTO NUMÉRICO E CATEGÓRICO + AUDITORIA (com correções)
# ================================
import numpy as np  # Importação de bibliotecas
import pandas as pd  # Importação de bibliotecas
import matplotlib.pyplot as plt  # Importação de bibliotecas

from sklearn.pipeline import Pipeline  # Criação de pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer

# -------------------------------------------------
# 1) NUMÉRICAS: age_* com indicador de faltantes
# -------------------------------------------------
age_cols = [  # Definição de colunas
    "age_first_funding_year", 
    "age_last_funding_year",
    "age_first_milestone_year", 
    "age_last_milestone_year"
]

age_pipeline = Pipeline(steps=[  # Criação de pipeline
    ("imputer", SimpleImputer(strategy="median", add_indicator=True))
])

# (visual opcional)
X_age = age_pipeline.fit_transform(X[age_cols])  # Definição de colunas
feat_names_age = age_pipeline.named_steps["imputer"].get_feature_names_out(age_cols)  # Definição de colunas
display(pd.DataFrame(X_age, columns=feat_names_age, index=X.index).head(10))


# -------------------------------------------------
# 2) NUMÉRICAS: outras (sem funding_total_usd)
# -------------------------------------------------
other_num_cols = [  # Definição de colunas
    "relationships", "funding_rounds", 
    "milestones", "avg_participants"
]

other_num_pipeline = Pipeline(steps=[  # Criação de pipeline
    ("imputer", SimpleImputer(strategy="median"))
])

# (visual opcional)
X_other_num = other_num_pipeline.fit_transform(X[other_num_cols])  # Definição de colunas
display(pd.DataFrame(X_other_num, columns=other_num_cols, index=X.index).head(10))  # Definição de colunas


# -------------------------------------------------
# 3) NUMÉRICA ESPECIAL: funding_total_usd (winsor 10–90% + RobustScaler)
# -------------------------------------------------
class QuantileClipper(BaseEstimator, TransformerMixin):
    def __init__(self, lower=0.01, upper=0.99):
        self.lower = lower
        self.upper = upper
        self.low_ = None
        self.high_ = None
    def fit(self, X, y=None):  # Treinamento do modelo
        X_ = np.asarray(X, dtype=float).reshape(-1, 1)
        self.low_  = np.nanpercentile(X_, self.lower * 100)
        self.high_ = np.nanpercentile(X_, self.upper * 100)
        return self
    def transform(self, X):
        X_ = np.asarray(X, dtype=float).reshape(-1, 1)
        return np.clip(X_, self.low_, self.high_)
    def get_feature_names_out(self, input_features=None):
        return np.asarray(input_features) if input_features is not None else np.array(["x0"])

funding_col = ["funding_total_usd"]
funding_pipeline = Pipeline(steps=[  # Criação de pipeline
    ("imputer", SimpleImputer(strategy="median")),
    ("clipper", QuantileClipper(lower=0.10, upper=0.90)),  # winsor mais agressivo
    ("scaler", RobustScaler())
])

# (auditoria opcional do funding antes/depois — fora do ColumnTransformer)
X_funding_before = X[funding_col].copy()
X_funding_after  = funding_pipeline.fit_transform(X_funding_before)
X_funding_after_df = pd.DataFrame(X_funding_after, columns=[f"{funding_col[0]}_winsor_robust"], index=X.index)
print("=== funding_total_usd (ANTES) ===")  
display(X_funding_before.describe(percentiles=[.01,.05,.5,.95,.99]).T)
print("=== funding_total_usd_winsor_robust (DEPOIS) ===")  
display(X_funding_after_df.describe(percentiles=[.01,.05,.5,.95,.99]).T)

def iqr_outlier_pct(s):
    s = pd.Series(s).dropna()
    if s.empty:
        return 0.0
    q1, q3 = np.percentile(s, [25, 75])
    iqr = q3 - q1
    if iqr == 0:
        return 0.0
    lo, hi = q1 - 1.5*iqr, q3 + 1.5*iqr
    return ((s < lo) | (s > hi)).mean() * 100

pct_before = iqr_outlier_pct(X_funding_before[funding_col[0]])
pct_after  = iqr_outlier_pct(X_funding_after_df.iloc[:, 0])
print(f"% de outliers (IQR) ANTES:  {pct_before:.2f}%")  
print(f"% de outliers (IQR) DEPOIS: {pct_after:.2f}%")  

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.boxplot(X_funding_before[funding_col[0]].dropna().values); plt.title("Antes: funding_total_usd")
plt.subplot(1,2,2); plt.boxplot(X_funding_after_df.iloc[:,0].dropna().values);   plt.title("Depois: funding_total_usd_winsor_robust")
plt.tight_layout(); plt.show()


# -------------------------------------------------
# 4) BINÁRIAS: garantia de 0 na ausência
# -------------------------------------------------
binary_pipeline = Pipeline(steps=[  # Criação de pipeline
    ("imputer", SimpleImputer(strategy="constant", fill_value=0))
])


# -------------------------------------------------
# 5) CATEGÓRICAS: estratégia para category_code (freq OU ohe_topn)
# -------------------------------------------------
from sklearn.base import BaseEstimator, TransformerMixin

# Diagnóstico rápido de cardinalidade
if "category_code" in X.columns:
    print("Cardinalidade de category_code:", X["category_code"].nunique())  

CAT_STRATEGY = "freq"   # mude para "ohe_topn" para Top-N + OTHER + One-Hot
TOP_N = 20              # usado quando CAT_STRATEGY="ohe_topn"

# separa category_code de eventuais outras categóricas
high_card_cat = [c for c in cat_cols if c == "category_code"]  # Definição de colunas
other_cat_cols = [c for c in cat_cols if c != "category_code"]  # Definição de colunas

class FrequencyEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):  # Treinamento do modelo
        s = (X.iloc[:,0] if isinstance(X, pd.DataFrame) else pd.Series(np.asarray(X).ravel())).astype(str)
        self.colname_ = X.columns[0] if isinstance(X, pd.DataFrame) else "cat"
        self.freq_ = s.value_counts(normalize=True).to_dict()
        return self
    def transform(self, X):
        s = (X.iloc[:,0] if isinstance(X, pd.DataFrame) else pd.Series(np.asarray(X).ravel())).astype(str)
        return s.map(self.freq_).fillna(0.0).to_numpy().reshape(-1,1)
    def get_feature_names_out(self, input_features=None):
        return np.array([f"{self.colname_}_freq"])

class TopNCategoryGrouper(BaseEstimator, TransformerMixin):
    def __init__(self, top_n=20): self.top_n = top_n
    def fit(self, X, y=None):  # Treinamento do modelo
        s = (X.iloc[:,0] if isinstance(X, pd.DataFrame) else pd.Series(np.asarray(X).ravel())).astype(str)
        self.colname_ = X.columns[0] if isinstance(X, pd.DataFrame) else "cat"
        self.keep_ = set(s.value_counts().head(self.top_n).index)
        return self
    def transform(self, X):
        s = (X.iloc[:,0] if isinstance(X, pd.DataFrame) else pd.Series(np.asarray(X).ravel())).astype(str)
        return s.where(s.isin(self.keep_), other="OTHER").to_frame()
    def get_feature_names_out(self, input_features=None):
        return np.array([self.colname_])

# pipelines categóricas
transformers_cat = []

# Outras categóricas (se houver) -> OneHot padrão
if other_cat_cols:
    cat_other_pipeline = Pipeline(steps=[  # Criação de pipeline
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
    ])
    transformers_cat.append(("cat_other", cat_other_pipeline, other_cat_cols))

# category_code com estratégia escolhida
if high_card_cat:
    if CAT_STRATEGY == "freq":
        cat_freq_pipeline = Pipeline(steps=[  # Criação de pipeline
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("freq", FrequencyEncoder())
        ])
        transformers_cat.append(("cat_freq", cat_freq_pipeline, high_card_cat))
    elif CAT_STRATEGY == "ohe_topn":
        cat_topn_pipeline = Pipeline(steps=[  # Criação de pipeline
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("topn", TopNCategoryGrouper(top_n=TOP_N)),
            ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
        ])
        transformers_cat.append(("cat_topn", cat_topn_pipeline, high_card_cat))
    else:
        raise ValueError("CAT_STRATEGY inválida: use 'freq' ou 'ohe_topn'.")


# -------------------------------------------------
# 6) ColumnTransformer final
# -------------------------------------------------
cat_pipeline = Pipeline(steps=[  # Criação de pipeline
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])  # este fica para as DEMAIS categóricas (caso não usemos transformers_cat); mantido por compatibilidade

transformers = []
if age_cols:
    transformers.append(("age", age_pipeline, age_cols))
if other_num_cols:
    transformers.append(("num", other_num_pipeline, other_num_cols))
if funding_col:
    transformers.append(("funding", funding_pipeline, funding_col))
if binary_cols:
    transformers.append(("bin", binary_pipeline, binary_cols))
if cat_cols:
    # Separamos category_code em transformers_cat; se não houver outras categóricas, esse cat_pipeline pode ficar sem efeito
    # A ordem aqui importa pouco; o essencial é incluir transformers_cat antes do ColumnTransformer
    pass

# adiciona blocos categóricos específicos (category_code e outras)
transformers.extend(transformers_cat)

preprocess = ColumnTransformer(
    transformers=transformers,
    remainder="drop"
)

# Fit + Transform
X_pp = preprocess.fit_transform(X)
print("Shape antes:", X.shape)  
print("Shape depois:", X_pp.shape)  

# Reconstrução do DataFrame pós-pipeline + auditoria de nulos
feat_names = preprocess.get_feature_names_out()
X_pp_df = pd.DataFrame(X_pp, columns=feat_names, index=X.index)

cols_excluir = [c for c in ["labels", "id"] if c in df_train.columns]  # Definição de colunas
miss_before = df_train.drop(columns=cols_excluir).isna().sum()  # Definição de colunas
miss_after  = X_pp_df.isna().sum()

print("=== Nulos por coluna (antes) — top 15 ===")  
display(miss_before.sort_values(ascending=False).head(15))
print("=== Nulos por coluna (depois) — top 15 ===")  
display(miss_after.sort_values(ascending=False).head(15))
print("Nulos totais antes:", int(miss_before.sum()))  
print("Nulos totais depois:", int(miss_after.sum()))  
```

### 05. Treinamento do Modelo + Validação da Importância de Features

```python
# ================================
# Feature importance (Permutation) com CV, pós-pipeline
# ================================
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.inspection import permutation_importance
from sklearn.ensemble import RandomForestClassifier

# 1) Dados
y_train = df_train["labels"].copy()
X_train = df_train.drop(columns=["labels", "id"], errors="ignore")

# 2) Pipeline simples: preprocess + modelo base (rápido)
rf_base = RandomForestClassifier(
    n_estimators=400,
    max_depth=None,
    n_jobs=-1,
    random_state=42,
    class_weight="balanced_subsample"  # ajuda se houver desbalanceamento
)

from sklearn.pipeline import Pipeline as SkPipeline  # Criação de pipeline
pipe_base = SkPipeline(steps=[  # Criação de pipeline
    ("preprocess", preprocess),
    ("clf", rf_base),
])

# 3) Treina pipeline em todo o treino (rápido)
pipe_base.fit(X_train, y_train)  # Treinamento do modelo

# 3.1) Transformar X pelo preprocess já treinado
pre = pipe_base.named_steps["preprocess"]
X_tr = pre.transform(X_train)                         # matriz após OneHot, winsor, etc.
feat_names = pre.get_feature_names_out()              # nomes EXPANDIDOS corretos

# 3.2) Estimador final
clf = pipe_base.named_steps["clf"]

# 4) Permutation importance no espaço TRANSFORMADO
from sklearn.inspection import permutation_importance
pi = permutation_importance(
    clf,
    X_tr, y_train,
    n_repeats=8,
    random_state=42,
    n_jobs=-1,
    scoring="accuracy"
)

# 5) Monta DataFrame de importâncias (agora com mesmo comprimento)
imp_df = pd.DataFrame({
    "feature": feat_names,
    "importance_mean": pi.importances_mean,
    "importance_std":  pi.importances_std
}).sort_values("importance_mean", ascending=False).reset_index(drop=True)

display(imp_df.head(20))

# 6) Visual: Top 30
top_k = 30
plot_df = imp_df.head(top_k).iloc[::-1]
plt.figure(figsize=(8, max(6, top_k*0.28)))
plt.barh(plot_df["feature"], plot_df["importance_mean"])
plt.xlabel("Permutation Importance (Δ Accuracy)")
plt.title(f"Top {top_k} features (Permutation Importance)")
plt.tight_layout()
plt.show()
```

### 06. Ajuste fino de Hiperparâmetros -> Seleção de features + RandomizedSearchCV (3 folds)

```python

import numpy as np  
import pandas as pd  
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline as SkPipeline  # Criação de pipeline
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    HistGradientBoostingClassifier,
    VotingClassifier
)

# ------------------ Guards básicos ------------------
# 1) Dados e preprocess precisam existir
if 'df_train' not in globals():
    df_train = pd.read_csv("./database/train.csv")
if 'df_test' not in globals():
    df_test = pd.read_csv("./database/test.csv")
assert 'preprocess' in globals(), "Rode a célula de PRÉ-PROCESSAMENTO (Célula 9) antes."

# 2) Célula A (Permutation Importance) deve ter sido rodada
assert 'imp_df' in globals() and 'pipe_base' in globals(), \
    "Rode a Célula A (Permutation Importance) antes para criar 'imp_df' e 'pipe_base'."

y_train = df_train["labels"].copy()
X_train = df_train.drop(columns=["labels","id"], errors="ignore")

# ------------------ Seletor robusto por nome ------------------
class SelectByName(BaseEstimator, TransformerMixin):

    def __init__(self, features_keep=None, names_all=None):
        self.features_keep = features_keep   
        self.names_all = names_all         

    def fit(self, X, y=None):  # Treinamento do modelo
        features_keep = list(self.features_keep)
        names_all = list(self.names_all)
        name_to_idx = {n: i for i, n in enumerate(names_all)}
        self.idx_ = [name_to_idx[n] for n in features_keep if n in name_to_idx]
        if len(self.idx_) == 0:
            raise ValueError(
                "Nenhuma feature mantida foi encontrada no snapshot de nomes. "
                "Verifique 'features_keep' vs 'names_all'."
            )
        self._names_all_ = names_all
        return self

    def transform(self, X):
        return X[:, self.idx_]

    def get_feature_names_out(self, input_features=None):
        return np.array([self._names_all_[i] for i in self.idx_])

# ------------------ Corte de features ------------------
mask_keep = imp_df["importance_mean"] > 0.0


features_keep = imp_df.loc[mask_keep, "feature"].tolist()
if len(features_keep) == 0:
    features_keep = imp_df.head(min(50, len(imp_df)))["feature"].tolist()

print(f"Total de features: {len(imp_df)} | Mantidas: {len(features_keep)}")  

# Snapshot dos nomes do preprocess (já fitado na Célula A dentro do pipe_base)
names_snapshot = pipe_base.named_steps["preprocess"].get_feature_names_out().tolist()
selector = SelectByName(features_keep=features_keep, names_all=names_snapshot)

preprocess_reduced = SkPipeline(steps=[  # Criação de pipeline
    ("preprocess", preprocess),   # será (re)fitado no search
    ("select", selector)          # usa apenas listas clonáveis
])

# Checagem rápida do shape reduzido
X_sel = preprocess_reduced.fit_transform(X_train, y_train)
print("Shape reduzido pós-seleção:", X_sel.shape)  

# ------------------ Ensemble + RandomizedSearch (rápido) ------------------
rf = RandomForestClassifier(
    n_estimators=300, max_depth=None,
    n_jobs=-1, random_state=42, class_weight="balanced_subsample"
)
et = ExtraTreesClassifier(
    n_estimators=300, max_depth=None,
    n_jobs=-1, random_state=42, class_weight="balanced_subsample"
)
hgb = HistGradientBoostingClassifier(random_state=42)

voter = VotingClassifier(
    estimators=[("rf", rf), ("et", et), ("hgb", hgb)],
    voting="soft"
)

pipe_ens_reduced = SkPipeline(steps=[  # Criação de pipeline
    ("prep", preprocess_reduced),
    ("clf", voter)
])

param_dist = {
    "clf__weights": [(1,1,1), (2,1,2), (1,2,2), (2,2,1), (1,1,2)],
    "clf__rf__min_samples_leaf": list(range(1, 5)),                 # ~ randint(1,5)
    "clf__rf__max_features": ["sqrt", "log2", None],
    "clf__et__min_samples_leaf": list(range(1, 5)),                 # ~ randint(1,5)
    "clf__et__max_features": ["sqrt", "log2", None],
    "clf__hgb__learning_rate": list(np.geomspace(0.02, 0.12, num=10)),   # ~ loguniform(0.02, 0.12)
    "clf__hgb__max_leaf_nodes": list(range(25, 131, 5)),                 # ~ randint(25,130)
    "clf__hgb__l2_regularization": list(np.geomspace(1e-4, 1e-1, num=8)) # ~ loguniform(1e-4, 1e-1)
}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
scoring = {"accuracy":"accuracy", "recall_macro":"recall_macro", "f1_macro":"f1_macro"}

rand_search = RandomizedSearchCV(
    estimator=pipe_ens_reduced,
    param_distributions=param_dist,
    n_iter=40,
    scoring=scoring,
    refit="accuracy",
    cv=cv,
    n_jobs=-1,
    verbose=1,
    random_state=42
)

rand_search.fit(X_train, y_train)  # Treinamento do modelo

best_idx = rand_search.best_index_
print("[REDUCED/FAST] best params:", rand_search.best_params_)  
print("[REDUCED/FAST] CV accuracy:", rand_search.cv_results_["mean_test_accuracy"][best_idx])  
print("[REDUCED/FAST] CV recall_macro:", rand_search.cv_results_["mean_test_recall_macro"][best_idx])  
print("[REDUCED/FAST] CV f1_macro:", rand_search.cv_results_["mean_test_f1_macro"][best_idx])  

# expõe como 'grid_ens_reduced' para reutilizar nas C/D/E
grid_ens_reduced = rand_search
```

### 07. Relatório e Métricas do Modelo -> Avaliação OOF (CV) + relatório e matriz de confusão

```python

import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  

from sklearn.model_selection import StratifiedKFold, cross_val_predict  
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

y_train = df_train["labels"].copy()
X_train = df_train.drop(columns=["labels","id"], errors="ignore")

# usa o melhor estimador do search leve (definido como grid_ens_reduced)
best_model = grid_ens_reduced.best_estimator_

# Predições OOF (out-of-fold)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
y_pred_oof  = cross_val_predict(best_model, X_train, y_train, cv=cv, n_jobs=-1, method="predict")  
y_proba_oof = cross_val_predict(best_model, X_train, y_train, cv=cv, n_jobs=-1, method="predict_proba")  

# Relatório agregado
print("=== Classification report (OOF) ===")  # Impressão de informações
print(classification_report(y_train, y_pred_oof, digits=4))  # Impressão de informações

# Matriz de confusão
labels_sorted = sorted(y_train.unique())
cm = confusion_matrix(y_train, y_pred_oof, labels=labels_sorted)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels_sorted)
fig, ax = plt.subplots(figsize=(5.5,5.5))
disp.plot(ax=ax, cmap="Blues", values_format="d", colorbar=False)
plt.title("Confusion Matrix (OOF)")
plt.tight_layout()
plt.show()
```

### 08. Tuning de priors p/ maximizar ACC

```python
import numpy as np  
from sklearn.metrics import accuracy_score

# usa as probabilidades OOF calculadas na Célula C
assert 'y_proba_oof' in globals(), "Rode a Célula C primeiro para obter y_proba_oof."

classes_ = np.array(sorted(y_train.unique()))
grid_vals = np.array([0.8, 0.9, 1.0, 1.1, 1.2], dtype=float)

# ===== Produto cartesiano com NumPy (sem itertools) =====
# Gera todas as combinações de 'grid_vals' repetidas para cada classe
k = len(classes_)
# meshgrid -> empilha -> reshape para (n_combinacoes, k)
grid_mesh = np.stack(np.meshgrid(*([grid_vals] * k), indexing='ij'), axis=-1)
grid_combinations = grid_mesh.reshape(-1, k)

best_acc, best_w = -1.0, None
for w in grid_combinations:
    proba_adj = y_proba_oof * w  # ajusta prior por classe
    proba_adj = proba_adj / proba_adj.sum(axis=1, keepdims=True)  # normaliza por linha
    y_pred_adj = classes_[np.argmax(proba_adj, axis=1)]
    acc = accuracy_score(y_train, y_pred_adj)
    if acc > best_acc:
        best_acc, best_w = acc, w

best_weights = dict(zip(classes_, best_w))
print("Melhor ACC OOF com priors:", best_acc)  
print("Pesos por classe (na ordem das classes):", best_weights)  

def apply_priors(proba, classes, weights_dict):
    """Aplica pesos por classe às probabilidades e renormaliza por linha."""
    w = np.array([weights_dict.get(c, 1.0) for c in classes], dtype=float)
    proba_adj = proba * w
    proba_adj = proba_adj / proba_adj.sum(axis=1, keepdims=True)
    return proba_adj
```

### 09. Treino Final + Submissão

```python

import numpy as np  # Importação de bibliotecas
import pandas as pd  # Importação de bibliotecas

# Checks
assert 'grid_ens_reduced' in globals(), "Rode a célula B (search) antes."
assert 'best_weights' in globals(), "Rode a célula D (priors) antes."
assert 'df_train' in globals() and 'df_test' in globals(), "Carregue df_train/df_test."

# Dados completos
y_train = df_train["labels"].copy()
X_train = df_train.drop(columns=["labels","id"], errors="ignore")
X_test_full = df_test.drop(columns=["labels","id"], errors="ignore")

# Modelo vencedor do search
best_model = grid_ens_reduced.best_estimator_

# Treina no dataset completo
best_model.fit(X_train, y_train)  # Treinamento do modelo

# Predição no teste (probabilidades)
proba_test = best_model.predict_proba(X_test_full)  # Predição com o modelo
classes_model = best_model.named_steps["clf"].classes_

# Função de aplicação dos priors (mesma da célula D)
def apply_priors(proba, classes, weights_dict):
    w = np.array([weights_dict.get(c, 1.0) for c in classes], dtype=float)
    proba_adj = proba * w
    proba_adj = proba_adj / proba_adj.sum(axis=1, keepdims=True)
    return proba_adj

# Ajuste de priors e rótulos finais
weights_for_model = {c: best_weights.get(c, 1.0) for c in classes_model}
proba_test_adj = apply_priors(proba_test, classes_model, weights_for_model)
y_pred_final = classes_model[np.argmax(proba_test_adj, axis=1)]

# CSV de submissão
sub_path = "submissao_final.csv"
pd.DataFrame({"id": df_test["id"], "labels": y_pred_final}).to_csv(sub_path, index=False)
print(f"Submissão salva em: {sub_path}")  # Impressão de informações
```

## 1. Introdução
O objetivo deste trabalho foi construir um modelo preditivo capaz de estimar a probabilidade de sucesso de startups a partir de informações sobre funding, localização, características de produto e investidores.  

O desenvolvimento envolveu etapas de **limpeza e tratamento de dados**, **exploração e análise visual (EDA)**, **formulação de hipóteses**, **engenharia de features**, **seleção de variáveis**, **modelagem supervisionada** e **avaliação comparativa** de desempenho.

---

## 2. Hipóteses Formuladas
Com base na exploração dos dados, foram formuladas as seguintes hipóteses sobre os fatores de sucesso de startups:

1. **Redes de Relacionamento**  
   Startups com maior número de relacionamentos (`num_relationships`) apresentam maior chance de sucesso devido ao acesso facilitado a parceiros, investidores e clientes.

2. **Volume e Persistência de Investimentos**  
   Empresas com maior número de rodadas de investimento e funding total acumulado (`funding_total_usd`, `funding_rounds`) tendem a possuir mais recursos para crescer e, portanto, maiores chances de se consolidar.

3. **Localização Estratégica**  
   Startups em hubs tecnológicos (NY, MA, CA) apresentam maior probabilidade de sucesso, dado o acesso a capital de risco, infraestrutura e ecossistemas de inovação.

Essas hipóteses foram testadas empiricamente por meio da **EDA** e posteriormente confirmadas pela **importância das features** no modelo final.

---

## 3. Pré-processamento e Engenharia de Features

### 3.1. Tratamento de Valores Nulos
- **Idades de funding e milestone (`age_*`)**: foi aplicada imputação pela mediana, com criação de indicadores binários de ausência (`add_indicator=True`).
- **Variáveis numéricas simples (`relationships`, `funding_rounds`, `milestones`, `avg_participants`)**: foi realizada imputação pela mediana.
- **Variáveis binárias (localização e categorias one-hot)**: foi aplicada imputação constante (0).
- **Variável categórica `category_code`**:  
  - Foi aplicado **frequency encoding** para reduzir a cardinalidade (34 categorias).  
  - Também foi considerada a abordagem *top-N + "Outros"*, como alternativa para reduzir sparsidade sem perda de relevância estatística.

### 3.2. Tratamento de Outliers
- A variável **`funding_total_usd`** apresentava forte assimetria e caudas longas.  
- Foi aplicada **winsorização entre 10% e 90% (QuantileClipper)** para reduzir o impacto de outliers extremos.  
- Após o clipping, foi utilizado o **RobustScaler** para normalização, preservando robustez contra valores atípicos.  
- Resultado: redução de **7,74% para 0,0%** de outliers detectados via IQR.

### 3.3. Seleção Automática de Features
- Foi utilizada a técnica de **Permutation Importance** sobre o ensemble inicial.  
- Features de baixa relevância foram descartadas para reduzir dimensionalidade e risco de overfitting.  
- As principais variáveis retidas incluíram:  
  - `num__relationships`  
  - `age__age_last_milestone_year`  
  - `num__avg_participants`  
  - `funding__funding_total_usd`  
  - `num__funding_rounds`

---

## 4. Modelagem Supervisionada

### 4.1. Escolha do Modelo
- Foram avaliados modelos individuais (RandomForest, ExtraTrees, HistGradientBoosting).
- O melhor desempenho foi obtido por meio de um **Voting Ensemble (soft voting)** que combinou os três classificadores:  
  - RandomForest: robusto a variância, utilizado como baseline.  
  - ExtraTrees: adicionou diversidade com maior aleatoriedade.  
  - HistGradientBoosting: capturou interações complexas em dados tabulares.

### 4.2. Ajuste de Hiperparâmetros
- Foi aplicado **RandomizedSearchCV** (3 folds, 25 iterações) para acelerar a busca.  
- Os espaços de busca incluíram:
  - `min_samples_leaf`, `max_features` para RF/ET.
  - `learning_rate`, `max_leaf_nodes`, `l2_regularization` para HGB.
  - Pesos do ensemble (`clf__weights`) para balancear contribuições dos modelos.
- Foi realizado ajuste de **priors de classe** no pós-processamento, com o objetivo de controlar o equilíbrio entre recall e precisão.
